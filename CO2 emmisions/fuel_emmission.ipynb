#Model 1 with 6 independent variables


```python
import statsmodels.api as sm

y=num_df['CO2EMISSIONS']
X=num_df.drop(columns=['CO2EMISSIONS'])

xc=sm.add_constant(X)
ols=sm.OLS(y,xc)
mod=ols.fit()
print(mod.summary())
```

#Model 1(b) with 5 independent variables


```python
X = num_df.drop(columns=['CO2EMISSIONS', 'FUELCONSUMPTION_COMB'])

xc1 = sm.add_constant(X)

ols1 = sm.OLS(y, xc1)
mod1 = ols1.fit()

print(mod1.summary())
```

#Model 2 with 4 independent variables


```python
y = num_df['CO2EMISSIONS']


X = num_df.drop(columns=['CO2EMISSIONS', 'FUELCONSUMPTION_CITY', 'FUELCONSUMPTION_HWY'])
xc2 = sm.add_constant(X)

ols2 = sm.OLS(y, xc2)
mod2 = ols2.fit()

print(mod2.summary())
```

#Model 3 with 3 independent variables


```python
y = num_df['CO2EMISSIONS']

X = num_df.drop(columns=['CO2EMISSIONS', 'FUELCONSUMPTION_CITY', 'FUELCONSUMPTION_HWY', 'FUELCONSUMPTION_COMB'])

xc3 = sm.add_constant(X)

ols3 = sm.OLS(y, xc3)
mod3 = ols3.fit()
print(mod3.summary())
```

```python
X = num_df.drop(columns=['CO2EMISSIONS','ENGINESIZE','CYLINDERS'])

y = num_df['CO2EMISSIONS']

xc4 = sm.add_constant(X)

ols4 = sm.OLS(y, xc4)
mod4 = ols4.fit()
print(mod4.summary())

```

```python

y=num_df['CO2EMISSIONS']
x1 = num_df.drop(columns=['CO2EMISSIONS', 'FUELCONSUMPTION_COMB'])
x2 = num_df.drop(columns=['CO2EMISSIONS', 'FUELCONSUMPTION_CITY', 'FUELCONSUMPTION_HWY'])
x3 = num_df.drop(columns=['CO2EMISSIONS', 'FUELCONSUMPTION_CITY', 'FUELCONSUMPTION_HWY', 'FUELCONSUMPTION_COMB'])
```

```python
from sklearn.model_selection import train_test_split
x1_train, x1_test, y1_train, y1_test = train_test_split(x1, y, test_size = 0.2, random_state = 42)
x2_train, x2_test, y2_train, y2_test = train_test_split(x2, y, test_size = 0.2, random_state = 42)
x3_train, x3_test, y3_train, y3_test = train_test_split(x3, y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
regressor1 = LinearRegression()
regressor2 = LinearRegression()
regressor3 = LinearRegression()
regressor1.fit(x1_train, y1_train)
regressor2.fit(x2_train, y2_train)
regressor3.fit(x3_train, y3_train)

```

```python
print(regressor1.coef_)
print(regressor1.intercept_)
print(regressor1.score(x1_test,y1_test))
```

```python
y1_pred = regressor1.predict(x1_test)
y2_pred = regressor2.predict(x2_test)
y3_pred = regressor3.predict(x3_test)
```
